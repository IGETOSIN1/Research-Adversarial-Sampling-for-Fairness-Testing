```python
def predict_input(path):
    #imagePath = 'C:\\NOISE\\DENOISE\\dog11.jpg'
    test_image = image.load_img(path, target_size = (32, 32)) 
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model.predict(test_image)
    classes = np.argmax(result, axis = 1)
    #print('\nPredicted Class:',np.argmax([result[0]]))
    prediction_class=np.argmax([result[0]])
    class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
    prediction_class=class_names[np.argmax([result[0]])]
    #print(path,"\t \t \t",prediction_class)
    #print(prediction_class)
    return prediction_class
```

```python
correct_classif_no=0
for i in range(0,1000):
    if i==0:
        j="0000"
    elif i<10:
        j="000"+str(i)
    elif i<100:
        j="00"+str(i)
    else:
        j="0"+str(i)
    url='C:\\NOISE\\dogsa\\'+str(i)+'.jpg'
    pre=predict_input(url)
    if pre=="dog":
        correct_classif_no=correct_classif_no+1
        print(correct_classif_no)    
    print(pre)
    print(correct_classif_no)
```

```python
correct_pred=0
for i in range(0,1000):
    if i==0:
        j="0000"
    elif i<10:
        j="000"+str(i)
    elif i<100:
        j="00"+str(i)
    else:
        j="0"+str(i)
    url='C:\\NOISE\\dogsa_clean\\'+str(i)+'.jpg'
    pre=predict_input(url)
    if pre !="dog":
        correct_pred=correct_pred+1
    print(correct_pred)
```
